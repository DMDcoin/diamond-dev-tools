```python
# Thesis:
# The parity nodes performance drasticly decreases,
# in the case that the network get's overloaded with Transactions.
# It is an effect that strenghens itself recursivly.
# if the nodes performance decreases just a little bit, 
# the queue gets longer, this decreases the networks performance further.
#
# Proof:
# The Network was able to handle up to 451 transactions per second,
# but had a breakdown with a transaction load of 538 transactions per second
#
#

# testalgo = 'hbbft'
# testrun = 'run3'
# testname = 'throughput1'

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# We use larger plot areas, adjust to your needs.
plt.rcParams["figure.figsize"] = (20, 10)
```

```python


#blocksFile = 'data/' + testalgo + '/' + testrun + '/' + testname + '/blocks.log.csv'

# blocksFile = 'data/hbbft/proofs/poolProtection/blocks.log.csv'

blocksFile = 'data/hbbft/proofs/poolProtection/tx350.csv'
blocksFile2 = 'data/hbbft/proofs/poolProtection/tx411.csv'
blocksFile3 = 'data/hbbft/proofs/poolProtection/tx451.csv'
blocksFile4 = 'data/hbbft/proofs/poolProtection/tx538.csv'


df1 = pd.read_csv(blocksFile)
df2 = pd.read_csv(blocksFile2)
df3 = pd.read_csv(blocksFile3)
df4 = pd.read_csv(blocksFile4)

def addCalcedColumns(df):
    df['blockTime'] =  df.timestamp - df.shift(1).timestamp
    df['txs'] = df.numberOfTransactions / df.blockTime


addCalcedColumns(df1)
addCalcedColumns(df2)
addCalcedColumns(df3)
addCalcedColumns(df4)

```

```python
# plot = plt.plot(df.blockTime)
plt.plot(df1.blockTime)
plt.plot(df2.blockTime)
plt.plot(df3.blockTime)
plt.plot(df4.blockTime)
```

```python


#blocksFile = 'data/' + testalgo + '/' + testrun + '/' + testname + '/blocks.log.csv'

# blocksFile = 'data/hbbft/proofs/poolProtection/blocks.log.csv'

blocksFile = 'data/hbbft/proofs/poolProtection/tx350.csv'
blocksFile2 = 'data/hbbft/proofs/poolProtection/tx411.csv'
blocksFile3 = 'data/hbbft/proofs/poolProtection/tx538.csv'


dfFixed = pd.read_csv(blocksFile)
dfFixed2 = pd.read_csv(blocksFile2)
dfFixed3 = pd.read_csv(blocksFile3)

dfFixed['blockTime'] =  dfFixed.timestamp - dfFixed.shift(1).timestamp
dfFixed['txs'] = dfFixed.numberOfTransactions / dfFixed.blockTime


dfFixed2['blockTime'] =  dfFixed2.timestamp - dfFixed2.shift(1).timestamp
dfFixed2['txs'] = dfFixed2.numberOfTransactions / dfFixed2.blockTime

dfFixed3['blockTime'] =  dfFixed3.timestamp - dfFixed3.shift(1).timestamp
dfFixed3['txs'] = dfFixed3.numberOfTransactions / dfFixed3.blockTime

```
